```python
import torch
torch.cuda.is_available()
```

```python
import os
from ultralytics import YOLO

DATASET_PATH = os.path.join(
    os.getcwd(), "datasets", "desmodus"
)

def export_yolo_model(model: YOLO):
    """Exporta el modelo yolo a tflite con Float16"""

    model.export(
        format="tflite",
        half=True,
        # int8=True,
        imgsz=320,
        workers=64,
        device="cuda",
        data=os.path.join(DATASET_PATH, "data.yaml"),
    )
```

### Load model

```python
pt_model = YOLO(os.path.join(
    os.getcwd(), "runs", "colab-train", "weights", "best.pt"))
pt_model
```

### Export .pt to .tflite

```python
export_yolo_model(pt_model)
```
