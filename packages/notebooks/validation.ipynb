### YOLO model

```python
trained_yolo = YOLO(
    os.path.join(os.getcwd(), "runs", "colab-train", "weights", "best.pt")
)
trained_yolo
```

```python
valid_images = os.listdir("datasets/desmodus/test/images")
valid_images
```

```python
for i in valid_images:
    print(i)
    trained_yolo.predict(
        source=os.path.join("datasets", "desmodus", "test", "images", i),
        conf=0.25,
        save=True
    )
```
